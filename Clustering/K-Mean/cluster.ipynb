```python
import numpy as np 
import pandas as pd 
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
%matplotlib inline

```

```python
clus_data = pd.read_csv("cluster_data.csv")
clus_data

X = clus_data.values[:,0:]
```

```python
plt.scatter(clus_data["Feature 1"] , clus_data["Feature 2"] )
plt.show
```

```python
clusterNum = 3
k_means = KMeans(init= "k-means++" , n_clusters= clusterNum , n_init= 12)
k_means.fit(X)

labels = k_means.labels_
clus_data["Clus_km"] = labels
clus_data.head()
```

```python
clus_data.groupby('Clus_km').mean
clus_data.head
```
