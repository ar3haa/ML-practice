```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("non_linear_regression_dataset.csv")
df.head(10)
```

```python
plt.figure(figsize=(8,5))

x_data,y_data = df["x"].values , df["y"].values 
plt.plot(x_data,y_data ,"ro")
plt.show()
```

```python
def model_func(x, a, b, c, d, e, f, g, h):
    return a*x**3 + b*x**2 + c*x + d + e*np.sin(f*x) + g*np.cos(h*x)
```

```python
from scipy.optimize import curve_fit


popt,pcov = curve_fit(model_func , x_data,y_data)
print(popt)
```

```python
plt.plot(x_data,y_data ,"ro")

x = x_data
y = model_func(x,*popt)
plt.plot(x,y)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y , y_data)

print("R-squared: ", r2)
```
